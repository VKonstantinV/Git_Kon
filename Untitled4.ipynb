```python
lower_p_price = df['price'].quantile(0.01)
upper_p_price = df['price'].quantile(0.99)
df_price_cleaned = df[(df['price'] >= lower_p_price) & (df['price'] <= upper_p_price)].copy()
print(f"В столбце price осталось {len(df_price_cleaned)} строк")
```

```python
lower_p_quantity = df['quantity'].quantile(0.01)
upper_p_quantity = df['quantity'].quantile(0.99)
df_quantity_cleaned = df[(df['quantity'] >= lower_p_quantity) & (df['quantity'] <= upper_p_quantity)].copy()
print(f"В столбце quantity осталось {len(df_quantity_cleaned)} строк")
```

```python
removed_price = df[~df.index.isin(df_price_cleaned.index)]
removed_quantity = df[~df.index.isin(df_quantity_cleaned.index)]

both_removed = pd.merge(removed_price, removed_quantity, how='inner', on=list(df.columns))
print(f"Количество строк, удаленных и по price, и по quantity: {len(both_removed)}")
```

```python
df_cleaned = df[
    (df['price'] >= lower_p_price) & (df['price'] <= upper_p_price) &
    (df['quantity'] >= lower_p_quantity) & (df['quantity'] <= upper_p_quantity)
].copy()

df_cleaned.describe()
```

### Выясним какие товары покупают чаще всего. Сезонность товаров

```python
def get_season(date):
    month = date.month
    if month in [12, 1, 2]:
        return 'зима'
    elif month in [3, 4, 5]:
        return 'весна'
    elif month in [6, 7, 8]:
        return 'лето'
    else:
        return 'осень'

df_cleaned['season'] = df_cleaned['date'].apply(get_season)
print(df_cleaned[['date', 'season']].head())
```

Анализ популярных товаров по сезонам

```python
top_products_season = (
    df_cleaned.groupby(['season', 'product'])['quantity']
    .sum()
    .reset_index()
)

top_products_season = (
    top_products_season.sort_values(['season', 'quantity'], ascending=[True, False])
    .groupby('season')
    .head())

print("Анализ популярных товаров по сезонам")
print(top_products_season.to_string())
```

### Категоризируем товары

```python
df['product'] = df['product'].str.lower()
```

```python
unique_products = df['product'].unique()
for product in unique_products:
    print(product)
```

```python
def categorize_product(product):
    product = product.lower()
    if ('растен' in product or 'бальзами' in product or 'бегони' in product or 'диам' in product or 'геран' in product or 'артемизи' in product or 'радерма' in product or
        'хризо' in product or 'суккулент' in product or 'эхеверия' in product or 'эвкалипт' in product or 'пеларгони' in product or 'адиантум' in product or
        'кампанула' in product or 'калатея' in product or 'нолина' in product or 'хлоро' in product or 'глоксин' in product or 'азали' in product or
        'хризанте' in product or 'гортенз' in product or 'иссоп' in product or 'аспарагу' in product or 'кашпо' in product or 'фиал' in product or
        'гербер' in product or 'мелис' in product or 'роза' in product):
        return 'Растения'
    elif 'рассад' in product or 'семен' in product or ' г ' in product or 'кассет' in product:
        return 'Семена'
    elif ('столов' in product or 'чайн' in product or 'тарел' in product or 'против' in product or 'крыш' in product or 'посуд' in product or 'круж' in product or
          'салат' in product or 'фужер' in product or 'бокал' in product or 'стакан' in product or 'стол' in product or 'кувш' in product or 'муляж' in product or
          'ков' in product or 'скатерт' in product or 'банк' in product or 'салф' in product or 'чехо' in product or 'штор' in product or 'подуш' in product or
          'искусств' in product or 'нож' in product or 'протв' in product or 'короб' in product):
        return 'Декор и посуда'
    elif ('гладильн' in product or 'утюг' in product or 'сушил' in product or 'таз' in product or 'щетк' in product or 'сумк' in product or 'набо' in product or
          'конте' in product or 'сверл' in product or 'лест' in product or 'стремя' in product or 'шваб' in product or 'вешалк' in product or 'ведро' in product or
          'перчат' in product or 'оцинк' in product or 'тележ' in product or 'телег' in product or 'вешалк' in product or 'корз' in product):
        return 'Хозтовары'
    else:
        return 'Прочее'

df_cleaned['category'] = df_cleaned['product'].apply(categorize_product)
```
